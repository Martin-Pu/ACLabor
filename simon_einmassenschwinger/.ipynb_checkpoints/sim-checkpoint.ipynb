```python
import params as st
from model import linSys
from feedforward import *

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

tSim = np.linspace(0, 2, 20001)

def updateOde(_):
    global res, uIn

    def sys(t, x, u, params):
        dx = linSys(t, x, u, params)

        return dx

    x0 = [0, 0]
    params = st.d, st.k, st.m, st.aMax
   
    uIn = stepOnOff()

    res = solve_ivp(sys,
                    [tSim[0], tSim[-1]],
                    x0,
                    t_eval=tSim,
                    args=(uIn, params))
```
